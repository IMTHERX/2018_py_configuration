```python
from netmiko import ConnectHandler
from datetime import datetime
#创建一个表示设备的字典。
cisco_rt1 = {
    'device_type': 'cisco_ios',
    'ip':'192.168.123.22',
    'username': 'cisco',
    'password': 'cisco',
    'port' : '22',          
    'secret': 'enable',     
    'verbose': False,      
}
cisco_sw = {
    'device_type': 'cisco_ios',
    'ip':'192.168.123.23',
    'username': 'cisco',
    'password': 'cisco',
    'port' : '22',          # optional, defaults to 22
    'secret': 'enable',     # optional, defaults to ''
    'verbose': False,       # optional, defaults to False
}

#所有设备的组
all_devices = [cisco_rt1,cisco_sw] 

#ssh
start_time = datetime.now()
print("开始时间:",start_time)

for a_device in all_devices:
    net_connect = ConnectHandler(**a_device)
    output = net_connect.send_command("sh ip int brief")    
    print ("\n\n>>>>>>>>>Device {0} <<<<<<<<<".format(a_device['device_type']))
    print (output)
    print (">>>>>>>>> End <<<<<<<<<")
    net_connect.disconnect()
end_time= datetime.now()
print("结束时间:",end_time)
print("总计耗时:",end_time - start_time)


#对于长时间运行的命令，请使用send_command_expect()
#send_command_expect 等待尾随提示符（或可选模式）
#net_connect.send_command_expect('write memory')

#断开ssh
#net_connect.disconnect()
```
